```python
#記号、数字、アルファベットを除去する関数を定義
def remove_symbol(text):
    removed = re.sub(r"[\W]+|[a-z]+|[0-9]+|[A-Z]+|ｗ", "", text)
    
    return removed
```

```python
def hyouka(num):
    replies = df["replies"][num]
    reply_list = replies.split(",")
    eva=0
    analyzer = oseti.Analyzer()
    count=0
    for reply in reply_list:
        removed = remove_symbol(reply)
        if removed=="":
            count +=1
            continue
        eva += analyzer.analyze(removed)[0]
    return eva/(len(reply_list)-count)
```

```python
#確認
for i in range(120):
    print("{}:{}".format(i,hyouka(i)))
```

```python
#polarityを定義
polarity = []
#各ツイートごとにリプライを評価
for replies in df["replies"]:
    #リプライを","で分割
    print(replies)
    replies = replies.split(",")
    #各リプライの極性の評価の合計値を定義
    eva = 0
    for rep in replies:
        #リプライごとに不要な記号を除去、その後おせちで評価してevaに加算
        removed = remove_symbol(rep)
        if removed=="":
            continue
        analyzer = oseti.Analyzer()
        print(analyzer.analyze(removed)[0])
        eva += analyzer.analyze(removed)[0]
    #polarityにツイートのリプライ評価（平均値）を代入
    polarity.append(eva/len(replies))
polarity
    
```

```python
#dfと結合
df["polarity"] = polarity
```

```python
#たぶんMeCabを通して単語に分解する関数を定義
def japanese_analyzer(string):
    result_list = []
    for line in tagger.parse(string).split("\n"):
        splited_line = line.split("\t")
        if len(splited_line) >= 2 and "名詞" in splited_line[1]:
            result_list.append(splited_line[0])
    return result_list
```
