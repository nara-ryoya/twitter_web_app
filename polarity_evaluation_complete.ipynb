```python
# 改変禁止（実行のみ）
#ライブラリのインポート
import numpy as np
import pandas as pd
import re
import oseti
import glob

# 不要な警告を非表示にする
import warnings
warnings.filterwarnings('ignore')
```

```python
#csvファイルの読み込み
# パスで指定したファイルの一覧をリスト形式で取得. （ここでは一階層下のtestファイル以下）
csv_files = glob.glob('tweet_data/*.csv')

#読み込むファイルのリストを表示
for a in csv_files:
    print(a)

#csvファイルの中身を追加していくリストを用意
data_list = []

#読み込むファイルのリストを走査
for file in csv_files:
    data_list.append(pd.read_csv(file))

#リストを全て行方向に結合
#axis=0:行方向に結合, sort
df = pd.concat(data_list, axis=0, sort=True)
```

```python
#記号、数字、アルファベットを除去する関数を定義
def remove_symbol(text):
    removed = re.sub(r"[\W]+|[a-z]+|[0-9]+|[A-Z]+|ｗ", "", text)
    
    return removed
```

```python
def hyouka(num):
    replies = df["replies"][num]
    reply_list = replies.split(",")
    eva=0
    analyzer = oseti.Analyzer()
    count=0
    for reply in reply_list:
        removed = remove_symbol(reply)
        if removed=="":
            count +=1
            continue
        eva += analyzer.analyze(removed)[0]
    return eva/(len(reply_list)-count)
```

```python
#確認
for i in range(120):
    print("{}:{}".format(i,hyouka(i)))
```

```python
#polarityを定義
polarity = []
#各ツイートごとにリプライを評価
for replies in df["replies"]:
    #リプライを","で分割
    print(replies)
    replies = replies.split(",")
    #各リプライの極性の評価の合計値を定義
    eva = 0
    for rep in replies:
        #リプライごとに不要な記号を除去、その後おせちで評価してevaに加算
        removed = remove_symbol(rep)
        if removed=="":
            continue
        analyzer = oseti.Analyzer()
        print(analyzer.analyze(removed)[0])
        eva += analyzer.analyze(removed)[0]
    #polarityにツイートのリプライ評価（平均値）を代入
    polarity.append(eva/len(replies))
polarity
    
```

```python
#dfと結合
df["polarity"] = polarity
```
